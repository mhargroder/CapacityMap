```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
inputfile = 'InputData/C104_NUMO.xlsx'
```

```python
node_df=pd.read_excel(inputfile, 'Nodes')
```

```python
node_list = node_df['Node'].values.tolist()
```

```python
G=nx.DiGraph()
```

```python
G.add_nodes_from(node_list)
```

Create a loop which interates over the data frame to create tuples and add to edges

```python
edge_df=pd.read_excel(inputfile, 'Edges')
```

```python
for index,row in edge_df.iterrows():
    G.add_edge(row['SupplyConstraint'] ,row['DemandConstraint'])
```

```python
plt.figure(figsize=(10,10))
nx.draw(G,with_labels = True, node_color = 'lightblue', node_size = 600, arrows = True,)
plt.savefig('NetworkMap.jpeg',bbox_inches='tight', pad_inches=2, transparent=True)
plt.show()
```
